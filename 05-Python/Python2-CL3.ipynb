```python
# 1)

"""
This definitely sprawled way more than I anticipated. Thought I was being clever but in hindsight don't think I was!
Likely some refactoring & commenting needed here.
"""

def cashpoint():
    withdrawalRequests = [10, 20, 40, 60, 80, 100]
    balance = 150

    def mainMenu(balance):
        print("Main Menu\nWelcome to Northern Frock\n1 - Display balance\n2 - Withdraw funds\n3 - Deposit funds\n9 - Return card")
        mainMenuOption = input('Enter an option:')
        if mainMenuOption == "1":
            print(f'Current Balance: £{balance}')
            return mainMenu(balance)
        elif mainMenuOption == "2":
            return withdrawalMenu(balance)
        elif mainMenuOption == "3":
            return depositMenu(balance)
        elif mainMenuOption == "9":
            return print("Thank you for using this ATM")
        else:
            print ("You must select a valid option")
            return mainMenu(balance)
    
    def withdrawalMenu(balance):
        print("Withdrawal\nPlease select withdrawal amount\n1 - £10\n2 - £20\n3 - £40\n4 - £60\n5 - £80\n6 - £100\n7 - Other amount\n8 - Return to main menu")
        withdrawalMenuOption = input('Enter an option:')

        try:
            withdrawalMenuOption = int(withdrawalMenuOption)
        except:
            print("Option must be an integer")
            return withdrawalMenu(balance)

        if not 0 < withdrawalMenuOption < 9:
            print ("You must select a valid option")
            return withdrawalMenu(balance)
        if withdrawalMenuOption == 8:
            return mainMenu(balance)
        elif withdrawalMenuOption == 7:
            customWithdrawal = input('Please enter a withdrawal amount')
            
            try:
                customWithdrawal = int(customWithdrawal)
            except:
                print("Withdrawal total must be an integer")
                return 
            if customWithdrawal % 10 != 0 or customWithdrawal < 1:
                print("Withdrawal total must be a positive multiple of 10")
                return withdrawalMenu(balance)
            if balance > customWithdrawal:
                balance -= customWithdrawal
                print(f'£{customWithdrawal} withdrawn')
                return mainMenu(balance)
        else:
            if balance >= withdrawalRequests[withdrawalMenuOption - 1]:
                balance -= withdrawalRequests[withdrawalMenuOption - 1]
                print(f'£{withdrawalRequests[withdrawalMenuOption - 1]} withdrawn')
                return mainMenu(balance)
            else:
                print("Insufficient funds")
                return withdrawalMenu(balance)

    def depositMenu(balance):
        print("Deposit")
        depositAmount = input('Enter an amount to deposit:')

        try:
            depositAmount = int(depositAmount)
        except:
            print("Deposit value must be an integer")
            return depositMenu(balance)
        
        if depositAmount < 1:
            print("Deposit value must be greater than 0")
            return depositMenu(balance)
        
        balance += depositAmount
        return mainMenu(balance)
            
    mainMenu(balance)

cashpoint()
```
